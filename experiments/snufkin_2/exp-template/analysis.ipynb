```python
from pathlib import Path

# Specify the root directory
root_dir = Path('/Users/ekvashyn/Code/mXf/magicXform-utils/experiments/snufkin_2/exp-template/out/')

# Initialize a dictionary to hold groups
groups = {}

# Walk through all directories in the root directory and its subdirectories
for dirpath in root_dir.glob('**'):
    if dirpath.is_dir():
        # Split the directory name by dots and extract the third element
        if dirpath.name != "out":
            third_param = dirpath.name.split('.')[2]
        else:
            third_param = ""
        
        # Add the directory to the appropriate group
        if third_param not in groups:
            groups[third_param] = []
        groups[third_param].append(dirpath)

print(f"group_name = {groups.items()}")

# Now, 'groups' contains directories grouped by the third parameter
for group_name, dirs in groups.items():
    print(f"Group: {group_name}")
    for dirpath in dirs:
        print(f" - {dirpath}")
```

```python
import pandas as pd

important_columns = ['index', 'status', 'time', 'SPACER_max_depth']

# Assuming 'groups' is the dictionary from the previous step
# where keys are the third parameters and values are lists of directories
# csv_0_path = "/Users/ekvashyn/Code/mXf/magicXform-utils/experiments/snufkin_2/exp-template/out/crown.enigma.benchmarks_combo.snufkin.cape.11_03_2024-t22-50-05/stats.csv"
# csv_0 = pd.read_csv(csv_0_path)[important_columns].rename(columns={'SPACER_max_depth': 'max_depth'})
# Initialize a dictionary to hold merged DataFrames for each group
merged_dfs = {}

for group_name, dirs in groups.items():
    # Initialize an empty DataFrame for this group
    merged_df = pd.DataFrame({"index": []})
    
    for dirpath in dirs:
        # Find all stats.csv files in the current directory
        csv_files = list(dirpath.glob('**/stats.csv'))
        
        for csv_file in csv_files:
            a = str(csv_file)
            # Read the CSV file into a DataFrame
            csv_name = a.split('/')[-2]
            fifth_param = csv_name.split('.')[4]
            df = pd.read_csv(csv_file)[important_columns].rename(columns={'SPACER_max_depth': f'max_depth_{fifth_param}',
                                                                          'status': f'status_{fifth_param}',
                                                                          'time': f'time_{fifth_param}'})
            
            # Merge the DataFrame with the existing one in the group
            # Assuming 'SPACER_max_depth' and 'time' are the columns to merge on
            # and they are unique within each file
            merged_df = pd.merge(merged_df, df, how='outer', on='index')
    
    # Store the merged DataFrame for this group
    merged_dfs[group_name] = merged_df

# Now, 'merged_dfs' contains merged DataFrames for each group
# You can save them to CSV files or further process them as needed
for group_name, df in merged_dfs.items():
    print(f"Group: {group_name}")
    filename = f"/Users/ekvashyn/Code/mXf/magicXform-utils/experiments/snufkin_2/exp-template/out_csvs/merged_{group_name}.csv"
    df = df[sorted(df.columns)]
    print(df.head()) 

    # Save the DataFrame to a CSV file
    df.to_csv(filename, index=False)
```
